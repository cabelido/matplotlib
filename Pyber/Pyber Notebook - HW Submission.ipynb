```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
citydf=pd.read_csv(city_data_to_load)
ridedf=pd.read_csv(ride_data_to_load)
citydf.head()
```

```python
ridedf.head()
```

```python
# Combine the data into a single dataset
mergeddf=pd.merge(ridedf,citydf, on="city",how="left")
# Display the data table for preview
mergeddf.head()
```

```python
mergeddf.count()
```

```python
#Filter urban, suburban and rural types and convert to dataframe with calculations for the bubble chart.
columns=["date",
         "city",
         "fare",
         "ride_id",
         "driver_count"]
         
urbandf=mergeddf.loc[mergeddf["type"]=="Urban",columns]
urbandf.head()
```

```python

urbanmod=urbandf.groupby(["city"])
ridecount=urbanmod["ride_id"].count()
avefare=urbanmod["fare"].mean()
drivercount=urbanmod["driver_count"].mean()
totalfare=urbanmod["fare"].sum()


urbandf2=pd.DataFrame({"Total Number of Rides Per City":ridecount,
                    "Average Fare $": avefare,
                    "Driver Count per City": drivercount})
urbandf2.head()
```

```python
suburbandf=mergeddf.loc[mergeddf["type"]=="Suburban",columns]
suburbandf.head()
```

```python
suburbanmod=suburbandf.groupby(["city"])
ridecount=suburbanmod["ride_id"].count()
avefare=suburbanmod["fare"].mean()
drivercount=suburbanmod["driver_count"].mean()

suburbandf2=pd.DataFrame({"Total Number of Rides Per City":ridecount,
                    "Average Fare $": avefare,
                    "Driver Count per City": drivercount})
suburbandf2.head()
```

```python
ruraldf=mergeddf.loc[mergeddf["type"]=="Rural",columns]
ruraldf.head()
```

```python
ruralmod=ruraldf.groupby(["city"])
ridecount=ruralmod["ride_id"].count()
avefare=ruralmod["fare"].mean()
drivercount=ruralmod["driver_count"].mean()

ruraldf2=pd.DataFrame({"Total Number of Rides Per City":ridecount,
                    "Average Fare $": avefare,
                    "Driver Count per City": drivercount})
ruraldf2.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
plt.scatter(urbandf2["Total Number of Rides Per City"], urbandf2["Average Fare $"], marker="o", facecolors="coral", edgecolors="black",
            s=(urbandf2["Driver Count per City"]*10), alpha=0.75,label="Urban")

plt.scatter(suburbandf2["Total Number of Rides Per City"], suburbandf2["Average Fare $"], marker="o", facecolors="lightblue", edgecolors="black",
            s=(suburbandf2["Driver Count per City"]*10), alpha=0.75,label="Suburban")

plt.scatter(ruraldf2["Total Number of Rides Per City"], ruraldf2["Average Fare $"], marker="o", facecolors="gold", edgecolors="black",
            s=ruraldf2["Driver Count per City"]*10, alpha=0.75,label="Rural")

# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid()

# Create a legend
lgnd = plt.legend(markerscale=.5, title="City Type")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,30,"Note:\n Circle size correlates with driver count per city")

# Save Figure
plt.savefig("../Images/Pyber_Ride_Sharing_Data2016.png",bbox_inches="tight")
```

## Total Fares by City Type

```python
# Calculate Type Percents
urbanfare = urbandf["fare"].sum()
suburbanfare = suburbandf["fare"].sum()
ruralfare = ruraldf["fare"].sum()

# Build Pie Chart
fig, ax = plt.subplots(figsize=(6,6)) 
ax.set_title("% of Total Fares by City Type")
explode = (0.1, 0, 0)

sizes = [urbanfare,suburbanfare,ruralfare]
labels = ["Urban","Suburban","Rural"]
colors = ["coral","lightblue","gold"]
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=240)
plt.axis("equal")

# Save Figure
plt.savefig("../Images/%_of_Total_Fares_by_City_Type.png",bbox_inches="tight")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urbanride = urbandf["ride_id"].count()
suburbanride = suburbandf["ride_id"].count()
ruralride = ruraldf["ride_id"].count()

# Build Pie Chart
fig, ax = plt.subplots(figsize=(6,6)) 
ax.set_title("% of Total Rides by City Type")

sizes = [urbanride,suburbanride,ruralride]
labels = ["Urban","Suburban","Rural"]
colors = ["coral","lightblue","gold"]
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=240)
plt.axis("equal")

# Save Figure
plt.savefig("../Images/%_of_Total_Rides_by_City_Type.png",bbox_inches="tight")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urbandriver=urbandf2["Driver Count per City"].sum()
suburbandriver=suburbandf2["Driver Count per City"].sum()
ruraldriver=ruraldf2["Driver Count per City"].sum()

# Build Pie Chart
fig, ax = plt.subplots(figsize=(6,6)) 
ax.set_title("Total Drivers by City Type")

sizes=[urbandriver,suburbandriver,ruraldriver]
labels=["Urban","Suburban","Rural"]
colors=["coral","lightblue","gold"]
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=240)
plt.axis("equal")

# Save Figure
plt.savefig("../Images/Total_Drivers_by_City_Type.png",bbox_inches="tight")
```
